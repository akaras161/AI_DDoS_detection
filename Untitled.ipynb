```python
Przepływ pakietów (np. 'Flow Packets/s', 'Fwd Packets/s', 'Total Fwd Packets') -> Ataki DDoS charakteryzują się gwałtownym wzrostem liczby pakietów wysyłanych i odbieranych.
Wielkość pakietów (np. 'Fwd Packet Length Max', 'Min Packet Length') -> Nietypowe długości pakietów mogą sygnalizować atak, np. bardzo małe lub duże pakiety.
Szybkość danych (np. 'Flow Bytes/s') -> W atakach DDoS ruch sieciowy gwałtownie wzrasta, więc przepływ danych na sekundę jest kluczowy.
Czas między pakietami (np. 'Flow IAT Mean') -> Regularność lub brak regularności w przychodzeniu pakietów jest istotnym wskaźnikiem.
Flagi TCP (np. 'SYN Flag Count', 'ACK Flag Count') -> SYN Flooding i inne ataki oparte na TCP mogą być łatwo wykrywane przez analizę flag TCP.
```

```python
source_ip_counter = Counter(dataframe[' Source IP'])
# Sortowanie wyników po wartościach liczników w porządku malejącym
sorted_source_ip = sorted(source_ip_counter.items(), key=lambda x: x[1], reverse=True)

print(sorted_source_ip)
```

```python
destination_ip_counter = Counter(dataframe[' Destination IP'])
# Sortowanie wyników po wartościach liczników w porządku malejącym
sorted_destination_ip = sorted(destination_ip_counter.items(), key=lambda x: x[1], reverse=True)

print(sorted_destination_ip)
```

```python
port_counter = Counter(dataframe[' Destination Port'])
# Sortowanie wyników po wartościach liczników w porządku malejącym
sorted_ports = sorted(port_counter.items(), key=lambda x: x[1], reverse=True)

print(sorted_ports)
```
